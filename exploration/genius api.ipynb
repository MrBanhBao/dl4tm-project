```python
def download_artists_lyrics(artist_list, num_songs, save_dir='../data/'):
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    
    for artist_item in artist_list:
        artist = genius.search_artist(artist_item, sort='popularity', max_songs=num_songs, get_full_info=True)
        artist.save_lyrics(os.path.join(save_dir, artist_item+'_lyrics.json'))
```

```python
artist_list = artists_80es + artists_90es + artists_00s + artists_10s
num_songs = 100
```

```python
download_artists_lyrics(artist_list, num_songs)
```
